```python
# Write afunction to calulate the sum of number from 1 to n
def myfunction(n):
  sum=0
  for x in range(1,n):
    sum=sum+x
  return sum
res=myfunction(10)
print(res)
```

```python
# Write a function that reurns the square of a number
def myfunction(n):
  return n*n
res=myfunction(10)
print(res)
```

```python
# Write a function that checks weather a number is even/odd
def myfunction(n):
  if(n%2==0):
    print("Even")
  else:
    print("Odd")
myfunction(10)
```

```python
# Write a function that returns the largest number from two numbers
def myfunction(x, y):
  if(x>y):
    print(x,"is largest")
  else:
    print(y,"is largest")
myfunction(12,78)
```

```python
#H7 Write a Python program to calculate the mean of a list of numbers.
l=[12,23,34,45,56,78,90]
sum=0
for x in range(0,7):
  sum=sum+l[x]
length=len(l)
mean=sum/length
print(mean)
```

```python
#H8 Classify Numbers into Even and Odd Lists
l=[12,23,34,45,56,78,90]
for x in range(0,7):
  if(l[x]%2==0):
    print("Even:",l[x])
  else:
    print("Odd:",l[x])
```
